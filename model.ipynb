```python
def unpickle(file):
    import pickle
    with open(file, 'rb') as fo:
        dict = pickle.load(fo, encoding='bytes')
    return dict
```

```python
## feature scale
from sklearn import preprocessing
scalar=preprocessing.StandardScaler()
scalar.fit(x)
x=scalar.transform(x)
x
```

```python
## use svm
from sklearn.svm import SVC
clf=SVC()
clf.fit(x,y)
```

```python
## collecting the test dataset
file='C://Users/HP/Desktop/cipher10/cifar-10-batches-py/test_batch'
dataset=unpickle(file)
x=dataset[b'data']
x=scalar.transform(x)
y=dataset[b'labels']
```

```python
## random forest score
clf.score(x,y)
```

```python
ypred=clf.predict(x)
```

```python
ypredlist=list(ypred)
```

```python
for i in range(len(ypredlist)):
    if(ypredlist[i]==0):
        ypredlist[i]='airplane'

    elif(ypredlist[i]==1):
        ypredlist[i]='automobile'

    elif(ypredlist[i]==2):
        ypredlist[i]='bird'
        
    elif(ypredlist[i]==3):
        ypredlist[i] = "cat"
        
    elif(ypredlist[i]==4):
        ypredlist[i]='deer'
        
    elif(ypredlist[i]==5):
        ypredlist[i]='dog'
        
    elif(ypredlist[i]==6):
        ypredlist[i]='frog'
        
    elif(ypredlist[i]==7):
        ypredlist[i]='horse'
        
    elif(ypredlist[i]==8):
        ypredlist[i]='ship'
        
    else:
        ypredlist[i]='truck'
```
